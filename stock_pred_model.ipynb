```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout
from datetime import date

def create_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))

    model.add(LSTM(units=60, activation='relu', return_sequences=True))
    model.add(Dropout(0.2))

    model.add(LSTM(units=80, activation='relu'))
    model.add(Dropout(0.2))

    model.add(Dense(units=1))

    model.compile(optimizer='adam', loss='mean_squared_error')

    return model


def train_model(stock_symbol, df):
    data = df[['Close']].values
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)
    train_size = int(len(scaled_data) * 0.9)
    train_data = scaled_data[:train_size]
    test_data = scaled_data[train_size:]

    # Prepare the training data
    x_train, y_train = [], []
    look_back = 60 
    for i in range(look_back, len(train_data)):
        x_train.append(train_data[i - look_back:i, 0])  
        y_train.append(train_data[i, 0])  

    x_train, y_train = np.array(x_train), np.array(y_train)
    x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
    model = create_model(input_shape=(x_train.shape[1], 1))
    model.fit(x_train, y_train, epochs=30, batch_size=32)
    x_test, y_test = [], []
    for i in range(look_back, len(test_data)):
        x_test.append(test_data[i - look_back:i, 0])
        y_test.append(test_data[i, 0])

    x_test, y_test = np.array(x_test), np.array(y_test)
    x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

    model.save(f"{stock_symbol}_model.h5")
    return model, scaler, x_test, y_test

def predict_next_7_days(stock_symbol, model, scaler, df):
    data = df[['Close']].values
    scaled_data = scaler.transform(data)

    inputs = scaled_data[-60:].reshape(1, -1)
    inputs = np.reshape(inputs, (inputs.shape[0], inputs.shape[1], 1))

    predictions = []
    for _ in range(7):
        pred = model.predict(inputs)
        predictions.append(pred[0][0])
        inputs = np.append(inputs[:, 1:, :], pred.reshape(1, 1, 1), axis=1)

    predictions = scaler.inverse_transform(np.array(predictions).reshape(-1, 1))
    return predictions

stocks = ['^NSEI', 'GOOG', 'TSLA']

models = {}
scalers = {}

for stock in stocks:
    try:
        print(f"Training model for {stock}...")
        current_date = date.today()
        df = yf.download(stock, start="2010-01-01", end=current_date)

        if not df.empty and 'Close' in df.columns:
            model, scaler, x_test, y_test = train_model(stock, df)
            models[stock] = model
            scalers[stock] = scaler

            y_pred = model.predict(x_test)
            y_pred = scaler.inverse_transform(y_pred)  
            y_test = scaler.inverse_transform(y_test.reshape(-1, 1))  

            plt.figure(figsize=(10, 5))
            plt.plot(y_test, label="Actual Test Data", marker='o', color='blue')
            plt.plot(y_pred, label="Predicted Test Data", marker='o', linestyle='dashed', color='red')
            plt.xlabel("Days")
            plt.ylabel("Stock Price")
            plt.title(f"Actual vs Predicted Stock Price for {stock}")
            plt.legend()
            plt.grid()
            plt.show()
        else:
            print(f"Insufficient data for {stock}. Skipping...")
    except Exception as e:
        print(f"Error downloading or training for {stock}: {e}")

user_stock = input("Enter a stock symbol (e.g., ^NSEI, AAPL, MSFT): ").upper()

if user_stock in models:
    model = models[user_stock]
    scaler = scalers[user_stock]
    df = yf.download(user_stock, start="2010-01-01", end=date.today())

    predictions = predict_next_7_days(user_stock, model, scaler, df)

    if predictions is not None:
        print(f"Predictions for the next 7 days for {user_stock}:")
        for i, pred in enumerate(predictions):
            print(f"Day {i + 1}: {pred[0]:.2f}")

        plt.plot(predictions, label="Predicted Stock Price", color='red')
        plt.title(f"7-Day Stock Price Prediction for {user_stock}")
        plt.xlabel('Days')
        plt.ylabel('Price')
        plt.legend()
        plt.show()
else:
    print(f"Stock {user_stock} is not in the trained models. Please train it first.")
```
